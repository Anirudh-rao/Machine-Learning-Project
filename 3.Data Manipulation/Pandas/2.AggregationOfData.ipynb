### Summary Statistics:

Summary Statistics is as the name suggests are numbers that summarize or tell ur about the dataset.

`df.mean()`: Will porvide the center of our dataset.

````python
print(df['Fuel_Price'].mean())
````

`df.median()`: Also provides the middle value of the dataset.

````python
print(df['CPI'].median())
````

`df.mode()`: Provides the most frequent values of the dataset.

````python
print(df['Unemployment'].mode())
````

`df.min()`: Provides the minimum value of the column

````python
print(df['Weekly_Sales'].min())
````

`df.max()`: Provides the maxiumum value of the column


````python
print(df['Weekly_Sales'].max())
````

`df.var()`: Provides the spread of the datapoint around their mean.


````python
print(df['Fuel_Price'].var())
````

`df.std()`: Provides the dispersion of data points relative to their mean.

````python
print(df['Fuel_Price'].std())
````

`df.sum()`: summation of the values of that column


````python
print(df['Fuel_Price'].sum())
````

`df.agg()`: Aggregation of value of a dataset.

````python
print(df['Fuel_Price'].agg('max'))
````

`df.cumsum()`: return the cumulative sum over a DataFrame or Series axis

````python
print(df.cumsum())
````

## Cumulative Statistics

`df.cummax()`: functions are used to calculate the cumulative maximum of values along a specified axis

````python
print(df.cummax())
````

`df.cumin()`:functions are used to calculate the cumulative minimum of values along a specified axis


````python
print(df.cummin())
````

`df.cumprod()`:The cumprod() function in pandas is used to calculate the cumulative product of values along a specified axis (index or columns) of a Series or DataFrame

````python
print(df.cumprod(skipna=False))
````

## Counting

We will use summary statistcs to count the value in a dataset.

`df.drop_duplicates[subset="<Column_name>"]`: This function will drop all duplicate values in the provided dataset and keep only the unique ones.


``` Note```: For passing multiple columns inside any function in pandas, Pandas always requires a list hence the given columns must be inside square brackets

````python
df.columns
````

````python
df.drop_duplicates(subset=['Weekly_Sales','Temperature'])
````

`df.value_counts()`: Counts the number of occurance of that particular value inside the dataset.

````python
df.value_counts()
````

````python
df.value_counts(sort=True)
````

`df.value_counts(normail=True)`: Will provide overall distribution of the column counts inside the dataset.

````python
df['Fuel_Price'].value_counts()
````

### Grouped Summary Statistics:

We can group two or more columns and hence summarize those using grouped summary statistics.

`df.groupby("Column_1")["Column_2"].mean()`

````python
df.groupby("CPI")['Temperature'].mean()
````

`df.groupby("Column_1")["Column_2"].agg[min,max,sum]`

````python
df.groupby("CPI")["Weekly_Sales"].agg[max]
````

## Pivot Tables:

`df.pivot_table(values='column_name1',index='color')`: here we can get a grouped summary staistics using pivot tables . By default, the pivot table functions takes in the mean. We can change that using `aggfunc` method and change the desired statistics.

1. `Columns argument`: creates the list of columns that we can pick as the view
2. `fill_value`: will fill in all the NAN values.
3. `Margin_value` = `True/False`: Will provide the mean weight of the summary statistics of rows and overall columns

df.groupby("CPI")["Weekly_Sales"].agg[max]
