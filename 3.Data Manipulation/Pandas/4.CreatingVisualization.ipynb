### Missing Values

Its important to handle missing data:

1. `df.isna()`: This function helps you find the number of `NaN` Not a number values in the dataset

```python
wallmartDf.isna()
```

```python
wallmartDf.isna().any()
```

```python
wallmartDf.isna().sum()
```

```python
wallmartDf.isna().sum().plot(kind="bar")
plt.show()
```

`df.dropna()`: This function removes all NA and NAN values from dataset.

```python
wallmartDf.dropna()
```

`df.fillna("value")`: Can be used to fill Na values with provided default value.

```python
wallmartDf.fillna(0)
```
