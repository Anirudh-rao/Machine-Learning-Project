### Explicit Indexes

`df.set_index("Column_name")`; Will help you set a particular column as an index.

```python
SetIndex = df.set_index('Surname')
print(SetIndex)
```

`df.reset_index()`: This function will help reset the index .

```python
SetIndex.reset_index()
print(SetIndex)
```

```python
df["Surname"].isin(["Walker"])
df
```

`df.sort_index()`: This function will help sort the index.

```python
df.sort_index()
df
```

### Slicing and subsetting with .loc and iloc

`df.loc`: Will slice the dataframe. Here we need to specify index values of the list.

```python
Surname = df.set_index(["Surname"]).sort_index()
Surname
```

```python
Surname.loc["Abbott":"Zuev"] 
```

As we can see from above example , just as we slice list  we must specify starting and ending value. `loc` function only performs the splitting on the index column, hence all inner columns do not work. To Slice inner index level we can follow the following method:

```python
Surname.loc[("Zubarev","5677"):("Zubarev","6940	")]
```

```python
Surname.loc[:,"Geography":"Gender"]
```

### Working with Pivot table

```python
wallmartDf = pd.read_csv('data/Walmart.csv')
wallmartDf.columns
wallmartDf.head(10)
```

```python
wallmartDf.columns
```

```python
Newdf = pd.pivot_table(wallmartDf,values=["Temperature","Weekly_Sales","Fuel_Price"],index="Date").sort_index()
Newdf
```

```python
Newdf.mean(axis="Fuel_Price")
```
