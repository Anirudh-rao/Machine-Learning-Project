### Dealing with Categorical Values:

Categorical values that values that hold Quanlitative data For ex: Country, colors etc. Machine learning requires these categorical values to be encoded into numbers.

We can create encoding for these categorical values. There are two approches:

1. `One-Hot Encoding`
2. `Dummy Encoding`

#### One-Hot Encoding: 

In this we  convert n categories into n features . We achieve this by using 

`df.get_dummies()` : This function takes in a dataframe and a list of categorical columns you want to be converted into one hot encoded columns and returns and updated dataframe with these columns included. We can provide prefix arguments that will make sure the columns are appropriately named.

```python
pd.get_dummies(df,columns=['Country'],
                prefix='C')
```

#### Dummy-Hot Encoding: 

In this , it created n-1 features for n categories ommitting the first category. Notice that this time there is no feature for france , the first category. In dummy encoding , the base value, france in this case , is encoded by the absence of all coutries as you can see on the last row here.Only differnece is that we use the drop_first argument and it is set to true.

```python
pd.get_dummies(df,columns=['Country'],
                drop_first=True, prefix='C')
```

#### One-Hot vs Dummies

1. One-hot Encoding: Explainable features.
2. Dummy Encoding : Neccessary information without duplication.

A Neccessary draw back for One-hot and dummy encoding is that these may explode the dataset to have many columns. This might cause fatigue in the machine learning leading to some cases of data fitting issues. Do note that its best to  encode categorical values on neccessary columns. 

```python
counts = df['Country'].value_counts()
print(counts)
```

Once the count is created, you can encode only those values that may contain count greater than certain value as show below.

```python
mask = df['Country'].isin(counts[counts<5].index)
df['Country'][mask] = 'Other'
```
