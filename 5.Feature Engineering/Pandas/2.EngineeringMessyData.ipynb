## Dealing with missing values:

When the missing values are not random, then we can use litwise deletion.
we can use the `df.dropna(subset=['Column_name'])`, this function will drop the rows that contain any Na values.

```python
df.dropna()
```

Drawbacks of litwise deletion:

1. It delete valid data points
2. Relies on randomness
3. Reduces information


The problem with deletion is as the name suggests , data gets removed and hence that data is lost. This can be of huge implications when we are building ML Models where data quality and data quantity is very much essential.

Instead of this we can replace the values with some preferd value
using `df.fillna(value='Default Value', inplace=True)`. This function will replace the values the `NA` values with the already provided values.

```python
df.fillna(value='None', inplace=True)
```

Similiar as seen above ,we can expand this thought to include other forms of data like:

`Categorical columns`: Replace missing values with the most common occuring value or with a string that flags missing values such as 'None'.

`Numeric Columns`: Replace the missing values with a sutiable value.

The suitable value may depend on what is required . In general cases we will make use of the central tendency i.e `mean` and `median`.Use of these central tendency functions should be carefully done as they may lead to data bias and incorrect error staistics.
