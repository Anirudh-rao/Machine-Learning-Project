```python
import pandas as pd
import scrapy as sp
from scrapy import Selector
```

```python
html = '''
<!DOCTYPE html>
<html>
<body>
    <h1>Random Text Generator</h1>
    <p id="randomTextDisplay">Your random text will appear here.</p>
    <button id="generateButton">Generate New Text</button>

    <script>
        const texts = [
            "Lorem ipsum dolor sit amet, consectetur adipiscing elit.",
            "Aenean commodo ligula eget dolor. Aenean massa.",
            "Cum sociis natoque penatibus et magnis dis parturient montes, nascetur ridiculus mus.",
            "Donec quam felis, ultricies nec, pellentesque eu, pretium quis, sem."
        ];

        function generateRandomText() {
            const randomIndex = Math.floor(Math.random() * texts.length);
            const randomText = texts[randomIndex];
            document.getElementById('randomTextDisplay').textContent = randomText;
        }

        document.getElementById('generateButton').addEventListener('click', generateRandomText);
    </script>
</body>
</html>

'''
```

```python
# Selector Selects all the html text
sel = Selector(text=html)
print(sel)
```

```python
# Selects all paragraph list
sel.xpath("//script")
```

```python
# Extract method extracs the value from selector list
sel.xpath("//p").extract
```
