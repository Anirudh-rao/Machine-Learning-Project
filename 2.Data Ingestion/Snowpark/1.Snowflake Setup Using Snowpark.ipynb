In this section , we will setup snowflake for our purpose. My goals for this section is to create a snowflake dwh of financial data from AlphaAdvantage.I am going to be creating three DWH:
1. **FX_DB**:Containing  all forex related data
2. **Crypto_DB**: Containing all Crypto related data.
3. **Commodities_DB**: Containing all Commodities Related databases.


This data will be used a learning and experimenting tool  for me to create a learn more about data engineering and data science and more about quantitative finance. 

```
# Import python packages
import pandas as pd

# We can also use Snowpark for our analyses!
from snowflake.snowpark.context import get_active_session
session = get_active_session()
```

```
create or replace warehouse SnowNotebook_Wh 
    warehouse_size = 'Small',
    generation = '2',
    auto_resume = True,
    auto_suspend = 60,
    comment = 'Warehouse Created for Notebook'
    ;

Use warehouse SnowNotebook_Wh

```

```
    
# Database Creation and Selection
session.sql('''create or replace database Finance_DB;''').collect()
session.sql('''Use database Finance_DB  ;''')
print(session.get_current_database())
```

```
# Schema Creation and Selection
session.sql('''create or replace schema ETL_DB.RAW;''').collect()
session.sql('''create or replace schema ETL_DB.CLEAN;''').collect()
session.sql('''use schema ETL_DB.RAW;''').collect()
print(session.get_current_schema())
```
