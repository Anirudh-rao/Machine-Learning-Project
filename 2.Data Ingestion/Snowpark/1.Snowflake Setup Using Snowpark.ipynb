```
# Import python packages
import pandas as pd

# We can also use Snowpark for our analyses!
from snowflake.snowpark.context import get_active_session
session = get_active_session()
```

```
create or replace warehouse SnowNotebook_Wh 
    warehouse_size = 'Small',
    generation = '2',
    auto_resume = True,
    auto_suspend = 60,
    comment = 'Warehouse Created for Notebook'
    ;

Use warehouse SnowNotebook_Wh

```

```
    
# Database Creation and Selection
session.sql('''create or replace database ETL_DB;''').collect()
session.sql('''Use database ETL_DB;''')
print(session.get_current_database())
```

```
# Schema Creation and Selection
session.sql('''create or replace schema ETL_DB.RAW;''').collect()
session.sql('''create or replace schema ETL_DB.CLEAN;''').collect()
session.sql('''use schema ETL_DB.RAW;''').collect()
print(session.get_current_schema())
```
