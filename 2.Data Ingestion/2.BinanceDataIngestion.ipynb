```python
Symbol = 'BTCUSDT'


#Format (year, month, day,hour, minute, second)
start_time = datetime.datetime(2025, 1, 1, 0, 0, 0)
end_time = datetime.datetime(2025, 1, 31, 0, 0, 0)
```

```python
klines = client.get_historical_klines(
    symbol = Symbol,
    interval = Client.KLINE_INTERVAL_5MINUTE,
    start_str=str(start_time), 
    end_str=str(end_time)
)
```

```python
df = pd.DataFrame(klines, columns=['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time', 'Quote Asset Volume', 'Number of Trades', 'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore'])
columns_convert = ['Open', 'High', 'Low', 'Close', 'Volume', 'Quote Asset Volume', 'Number of Trades', 'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume']

for col in columns_convert:
    df[col] = df[col].astype(float)

df
```
