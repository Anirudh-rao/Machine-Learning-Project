In this section we will cover data Ingestion using AlphaAdvantage API and Ingest Forex Daily Data. For more information follow this [Link](https://www.alphavantage.co/documentation/)

```python
import requests
import pandas as pd
import os

env = os.getenv('API')

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = f'https://www.alphavantage.co/query?function=FX_WEEKLY&from_symbol=EUR&to_symbol=USD&apikey={env}'
r = requests.get(url)
data = r.json()

print(type(data))
```

```python
df = pd.DataFrame.from_dict(data)
df.head(10)
```
