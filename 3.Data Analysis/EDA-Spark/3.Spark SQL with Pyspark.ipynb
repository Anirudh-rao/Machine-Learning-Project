## Resilient Distributed Datasets in Pyspark

RDD's are :
1. Distributed data collections across a cluster with automatic recovery from node failures.

2. Good for scale.

3. Immutable and can be transformed using operations like map(), filter() with actions like collect() or paralelize() to retrieve results or create RDD's

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('RDD Example').getOrCreate()

```

```python
df = spark.read.csv('/Volumes/learning/source/source/rows.csv', inferSchema=True)
```

```python
# Convert DF to RDD
RDD_df =  df.rdd

# Show the RDD contents using collect()
RDD_df.collect()
```
