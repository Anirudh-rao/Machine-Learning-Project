Lets convert prop_id columns from `object` into a `String`

## Training and Test Splits

We split the data into traiining and testing data to:


1. Reduce overfitting

2. Evaluate Performance on a holdout set

```python
df = pd.read_csv("data/Volunteering.csv")
print(df.head())
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split


X_train, X_test, y_testy_train, y_test = train_test_split(df, df, test_size=0.2, random_state=42)
```

```python
print(X_train.count())
print(X_test.count())
print(y_train.count())
print(y_train.count())
```

The most common mistake that can happen here is called `class imbalance`. The data is randomly split hence certain data points in the testing set may not be there in training and vice versa. This cause the model to have a bias and mispredict its predictions.To solve this we have some techniques like:

Startified Sampling:In this type of sampling , we take the account of all the types of classes in the dataset that we are predicting and then we split the data equally based on classes.

```python
X_train, X_test, y_testy_train, y_test = train_test_split(df, df, stratify=df, random_state=42)
```
