In this Section we will cover feature Engineering. Feature engineering is the process of creation of new features from existing ones. This helps improve performance and give insight into relationships between features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
voluenteer_df = pd.read_csv('data/Volunteering.csv')
voluenteer_df.head(10)
```

```python
voluenteer_df.columns
```

## Encoding Categorical Variables



```python
# Data Cleaning
voluenteer_df.fillna(value='0', inplace=True)
voluenteer_df.head(10)
```

```python
df = voluenteer_df
df['is_priority'].unique()
```

```python
df['is_priority_enc'] = df['is_priority'].apply(lambda val:1 if val=='y' else 0)
df[['is_priority_enc','is_priority']]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df2 = voluenteer_df

df2['is_priority_enc'] = le.fit_transform(df2['is_priority'])
df2[['is_priority_enc','is_priority']]
```

## One Hot Encoding:
One hot encoding is another method that encodes categorical variables with multiple values.


```python
print(df2['category_desc'])
```

```python
print(pd.get_dummies(df2['category_desc']))
```

## Engineering Numerical Features:


```python
print(df2['end_date_date'])
```

```python
df2['Date_Converted'] = pd.to_datetime(df2['end_date_date'])
df2['Month'] = df2['Date_Converted'].dt.month

print(df2[['Date_Converted','Month','end_date_date']])
```

## Engineering text features:

**Extractions**: Extracting data from text fields and then converting it into a features.


```python
import re

my_string = "Temperature:75.6F"

temp = re.search("\d+\.\d+",my_string)

print(float(temp.group(0)))
```
