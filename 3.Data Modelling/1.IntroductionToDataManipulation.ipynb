### Sorting and subsetting:

`df.sort_values(<'column_name'>, ascending=True/False)`: This will sort the dataset based on a particular column

```python
df.head(10)
```

```python
df.sort_values('Geography')
```

```python
df.sort_values('Geography', ascending=False)
```

```python
# Sorting by multiple variables
df.sort_values(['Geography','Surname'])
```

```python
df.sort_values(['Geography','Surname'], ascending=[True, False])
```

`df['<Column_Name'>]`: This is called subsetting a column from the dataframe.

```python
df['Surname']
```

`df[['<Column_Name1>',''<Column_Name2>'']]`: This is called subsetting multiple columns from the dataframe.

```python
df[['Geography','Surname']]
```

```python
# One more method
col_name = ['Geography','Surname']
df[col_name]
```

```python
# Filtering data using subsetting
df['CreditScore'] > 700
```

```python
#Filtering across the dataframe
df[df['CreditScore'] > 700]
```

```python
# Subestting based on Text
df[df['Geography'] == 'Spain']
```

```python
# Subsetting based on multiple conditions
isCredit = df['CreditScore'] > 700
isSpain = df['Geography'] == 'Spain'
df[isCredit &  isSpain]
```

```python
df[(df['Geography'] == 'Spain') & (df['CreditScore'] > 700)]
```

`isin()`: This method is used to check some logical condition from the dataframe.

```python
df['Geography'].isin(['France','Spain'])
```
