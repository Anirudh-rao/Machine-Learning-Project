## What is the center value of data?

1. **Mean**: Most common measure of central tendency its the total sum of n objects divided by n  number of objects.

2. **Median**: In short this is the middle value in a ordered dataset. This can be done by sorting and then taking the median.

3. **Mode** : The most common occurance of the data.Most useful for categorical variables.

## What is measure of Spread?

Spread is as the name suggests shows how spread or distribution of the data. There are some of the methods to measure spread. They are:


1. **Variance**: Average distance from each data point to the data's mean.

2. **Standar Deviation**: This is obtained by squaring the Variance.


3. **Mean Absolute Deviation**: Takes the absolute value of deviation to the mean.

Standard Deviation vs Mean Absolute Deviation:

1. Standard deviation squared distances, penilizing longer distances more than shorter distances.

2. Mean absoulte deviation penalized each distance equally.


**Quartiles/Percentiles**: This splits data into 4 equal parts.

**Interquartile Range**: This is the measure of 25% and 75% if the data.
