**3.Normal Distribution**: Normal distribution offten depicts a bell shaped curve. Some properties of normal distribution are:

1. The area under the curve is 1.
   
2. The curve never hits 0, even though it seems to be touching 0.

Its described by the `mean` and `standard deviation`.
When the mean = 0 and sd = 1 , then its called standard normal distribution.

### Central Limit Theorem:
The sampling distribution of a statistic becomes closed to normal distribution as the number of trail increases. For central limit theory to apply , the samples must be random and indepenedent.
