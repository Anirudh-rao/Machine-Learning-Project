## Design of Experiment

Experiments aims to answer: What is the effect of the treatement on the responses?

1. Treatement : Explanatory/independent variable.

2. Response : response/ dependent variable.


Controlled Experiement:

Participants are assigned by researchers to either treatement group or control group.

The Gold standard of experiment will use:

1. Randomized controlled trail

    1. Participatns are assigned to treatement/control randomly, not based on any other charectaristics.

    2. Choosing randomly helps ensure the groups are comparable

2. Placebo
   1. Resembles tretement but has not effect
   
   2. Participants will not know which group they are in

3. Double-Blind trail
    
    1. Person adminstering the treatment/running the study doesn't know whether the treatement is real or placebo

    2. Prevent bias in the response and/or analysis of results.
