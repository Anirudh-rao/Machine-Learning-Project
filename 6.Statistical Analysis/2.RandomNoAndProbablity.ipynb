## What are the chance?

Chance is the probability of something of hapenning.

```
P(Event) = Number of ways event can happen/Total number of possible outcomes

```
Probality is always between `0` and `1`.


### Independent events:

Two events are independent if the probablity of the second even isn't affected by the outcome of the first event.


### Dependent events:

Two events are depenedent  if the probablity of the second event is affected by the outcome of the first event.
